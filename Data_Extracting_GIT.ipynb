# Data Cleaning

```python
# 1. aggre_transcation

#Enter your full path
path_transaction = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/aggregated/transaction/country/india/state/"
agg_tran_list = os.listdir(path_transaction) #Returns a list of filenames in the specified directory.

columns1={"State":[], "Year":[], "Quarter":[],"Transaction_Type":[],
          "Transaction_Count":[], "Transaction_Amount":[]}

for state in agg_tran_list:
    current_state = path_transaction+state+"/"
    agg_year_list = os.listdir(current_state)
    
    for year in agg_year_list:
        current_year = current_state+year+"/"
        agg_file_list = os.listdir(current_year)
        
        for file in agg_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            aggre_tran=json.load(data)
            
            for i in aggre_tran["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]

                columns1["Transaction_Type"].append(name)
                columns1["Transaction_Count"].append(count)
                columns1["Transaction_Amount"].append(amount)
                
                columns1["State"].append(state)
                columns1["Year"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
     

aggre_transaction = pd.DataFrame(columns1)

aggre_transaction["State"] = aggre_transaction["State"].str.replace("-", " ").str.title()
aggre_transaction['State'] = aggre_transaction['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# 2. aggre_user

#Enter your full path
path_user = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/aggregated/user/country/india/state/"
agg_user_list=os.listdir(path_user)

columns2={"State":[], "Year":[], "Quarter":[],"Brand":[],
          "Brand_T_Count":[], "Brand_T_Percentage":[]}

for state in agg_user_list:
    current_state = path_user+state+"/"
    agg_year_list = os.listdir(current_state)
    
    for year in agg_year_list:
        current_year = current_state+year+"/"
        agg_file_list = os.listdir(current_year)
        
        for file in agg_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            aggre_user=json.load(data)
            
            try:
                for i in aggre_user["data"]["usersByDevice"]:
                    brand = i["brand"]
                    count = i["count"]
                    percentage = i["percentage"]

                    columns2["Brand"].append(brand)
                    columns2["Brand_T_Count"].append(count)
                    columns2["Brand_T_Percentage"].append(percentage)

                    columns2["State"].append(state)
                    columns2["Year"].append(year)
                    columns2["Quarter"].append(int(file.strip(".json")))
                    
            except:
                pass

aggre_user = pd.DataFrame(columns2)

aggre_user["State"] = aggre_user["State"].str.replace("-", " ").str.title()
aggre_user['State'] = aggre_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# 3. map_transaction

#Enter your full path
path_map_transaction = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/map/transaction/hover/country/india/state/"
map_tran_list = os.listdir(path_map_transaction)

columns3={"State":[], "Year":[], "Quarter":[],"District":[],"District_T_Count":[], "District_T_Amount":[]}

for state in map_tran_list:
    current_state = path_map_transaction+state+"/"
    map_year_list = os.listdir(current_state)
    
    for year in map_year_list:
        current_year = current_state+year+"/"
        map_file_list = os.listdir(current_year)
        
        for file in map_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            map_tran=json.load(data)
            
            for i in map_tran["data"]["hoverDataList"]:
                district=i["name"]
                count=i["metric"][0]["count"]
                amount=i["metric"][0]["amount"]

                columns3["District"].append(district)
                columns3["District_T_Count"].append(count)
                columns3["District_T_Amount"].append(amount)

                columns3["State"].append(state)
                columns3["Year"].append(year)
                columns3["Quarter"].append(int(file.strip(".json")))
        
map_transaction = pd.DataFrame(columns3)

map_transaction["State"] = map_transaction["State"].str.replace("-", " ").str.title()
map_transaction['State'] = map_transaction['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# 4. map_user

#Enter your full path
path_map_user = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path_map_user)

columns4={"State":[], "Year":[], "Quarter":[],"District":[],"District_Reg_User":[], "App_Opens":[]}

for state in map_user_list:
    current_state = path_map_user+state+"/"
    map_year_list = os.listdir(current_state)
    
    for year in map_year_list:
        current_year = current_state+year+"/"
        map_file_list = os.listdir(current_year)
        
        for file in map_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            map_user=json.load(data)
            
            for i in map_user["data"]["hoverData"].items():
                district=i[0]
                user=i[1]["registeredUsers"]
                appOpens=i[1]["appOpens"]

                columns4["District"].append(district)
                columns4["District_Reg_User"].append(user)
                columns4["App_Opens"].append(appOpens)

                columns4["State"].append(state)
                columns4["Year"].append(year)
                columns4["Quarter"].append(int(file.strip(".json")))
                
map_user = pd.DataFrame(columns4)

map_user["State"] = map_user["State"].str.replace("-", " ").str.title()
map_user['State'] = map_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# 5. top_transaction

#Enter your full path
path_top_tran = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/top/transaction/country/india/state/"
top_tran_list = os.listdir(path_top_tran)

columns5={"State":[], "Year":[], "Quarter":[],"Pincode":[],"Pincode_T_Count":[], "Pincode_T_Amount":[]}

for state in top_tran_list:
    current_state = path_top_tran+state+"/"
    top_year_list = os.listdir(current_state)
    
    for year in top_year_list:
        current_year = current_state+year+"/"
        top_file_list = os.listdir(current_year)
        
        for file in top_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            top_tran=json.load(data)
            
            for i in top_tran["data"]["pincodes"]:
                pincode = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]
                
                columns5["Pincode"].append(pincode)
                columns5["Pincode_T_Count"].append(count)
                columns5["Pincode_T_Amount"].append(amount)

                columns5["State"].append(state)
                columns5["Year"].append(year)
                columns5["Quarter"].append(int(file.strip(".json")))
                
top_transaction = pd.DataFrame(columns5)

top_transaction["State"] = top_transaction["State"].str.replace("-", " ").str.title()
top_transaction['State'] = top_transaction['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# 6. top_user

#Enter your full path
path_top_user = "........./Phonepe Pulse Data Visualization and Exploration/pulse/data/top/user/country/india/state/"
top_user_list = os.listdir(path_top_user)

columns6={"State":[], "Year":[], "Quarter":[],"Pincode":[],"Pincode_Reg_User":[]}

for state in top_user_list:
    current_state = path_top_user+state+"/"
    top_year_list = os.listdir(current_state)
    
    for year in top_year_list:
        current_year = current_state+year+"/"
        top_file_list = os.listdir(current_year)
        
        for file in top_file_list:
            current_file = current_year+file
            data = open(current_file, "r")
            
            top_user=json.load(data)
        
            for i in top_user["data"]["pincodes"]:
                pincode = i["name"]
                count = i["registeredUsers"]
                
                columns6["Pincode"].append(pincode)
                columns6["Pincode_Reg_User"].append(count)

                columns6["State"].append(state)
                columns6["Year"].append(year)
                columns6["Quarter"].append(int(file.strip(".json")))
                
top_user = pd.DataFrame(columns6)

top_user["State"] = top_user["State"].str.replace("-", " ").str.title()
top_user['State'] = top_user['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```
